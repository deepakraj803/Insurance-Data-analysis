comapring feature vs feature


```python
sns.pairplot(df, hue='smoker')  # Create pair plots to visualize relationships between features
plt.show()
```

```python
sns.lmplot(x='age', y='charges', data=df, hue='smoker')  # Create an implot to visualize the relationship between 'age' and 'charges' with 'smoker' as hue
plt.title('Age vs Charges with Smoker Hue')
```

Observation

1.Smokers (red crosses) have significantly higher charges than non-smokers (blue circles).

2.For smokers, charges increase steeply with age, showing a strong positive correlation.

3.For non-smokers, the increase in charges with age is more gradual.

4.The regression lines clearly show that age is a stronger predictor of charges for smokers than for non-smokers.
